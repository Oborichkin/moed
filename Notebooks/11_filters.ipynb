```python
import sys
from random import random
sys.path.append('../')

from moed.model import Sequence, Model
from moed.analysis import Analysis
from moed.proc import Proc, Filter

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
size = 64
dt = 0.001
fc = 100
```

```python
lpf = Filter.low_pass_filter(size, dt, fc)
dft_lpf = Analysis.dtf(lpf)
dft_lpf.plot()
lpf.plot()
```

```python
hpf = Filter.high_pass_filter(size, dt, fc)
dft_hpf = Analysis.dtf(hpf)
hpf.plot()
dft_hpf.plot()
```

```python
x = range(size)
fig, axes = plt.subplots(1, figsize=(15, 10))
axes.plot(x, dft_lpf.y, label="lpf")
axes.plot(x, dft_hpf.y, label="hpf")
axes.legend()
plt.show()
```
