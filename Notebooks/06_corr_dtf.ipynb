```python
_range = [x / 10 for x in range(250)]

signal = Model.harmonic(_range)
signal.plot()
```

```python
noise = Sequence.from_func(_range, lambda x: random()*3) + signal
noise2 = Sequence.from_func(_range, lambda x: random()*3) + signal
noise.plot()
```

```python
a = Analysis.auto_correlation(noise)
a.plot()
```

```python
c = Analysis.cross_correlation(noise, signal)
a.plot()
```
